```python
# ==========================================================
# DAY 22 - FINAL VALIDATION & DASHBOARD CHECK
# ==========================================================

import pandas as pd
from pathlib import Path

PROJECT_ROOT = Path().resolve().parent

INPUT_FOLDER = PROJECT_ROOT / "05_powerbi_data"
OUTPUT_FILE = PROJECT_ROOT / "05_powerbi_data/final_validated_dataset.csv"

df = pd.read_csv(INPUT_FOLDER / "dashboard_main.csv")

print("Dataset Loaded:", df.shape)

# ----------------------------------------------------------
# 1. CHECK FOR MISSING VALUES
# ----------------------------------------------------------

missing = df.isnull().sum()

print("\nMissing Values:\n", missing)

# Drop if any (strict cleaning)
df = df.dropna()

# ----------------------------------------------------------
# 2. DATA TYPE VALIDATION
# ----------------------------------------------------------

print("\nData Types:\n", df.dtypes)

# Ensure numeric columns are correct
numeric_cols = [
    "total_quantity", "avg_cost", "avg_defect",
    "efficiency", "reliability_score",
    "risk_score", "spoilage_rate",
    "performance_score"
]

for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors="coerce")

# ----------------------------------------------------------
# 3. KPI CROSS CHECK
# ----------------------------------------------------------

print("\nKPI VALIDATION:")

print("Total Vendors:", len(df))
print("Top Performers:", len(df[df["final_category"] == "Top Performer"]))
print("Critical Suppliers:", len(df[df["final_category"] == "Critical Risk"]))
print("Avg Cost:", df["avg_cost"].mean())
print("Avg Defect:", df["avg_defect"].mean())

# ----------------------------------------------------------
# 4. DUPLICATE CHECK
# ----------------------------------------------------------

duplicates = df.duplicated().sum()
print("\nDuplicate Rows:", duplicates)

df = df.drop_duplicates()

# ----------------------------------------------------------
# 5. RANGE VALIDATION (Sanity Checks)
# ----------------------------------------------------------

print("\nSanity Checks:")

print("Min Cost:", df["avg_cost"].min())
print("Max Cost:", df["avg_cost"].max())

print("Min Defect:", df["avg_defect"].min())
print("Max Defect:", df["avg_defect"].max())

print("Min Score:", df["performance_score"].min())
print("Max Score:", df["performance_score"].max())

# ----------------------------------------------------------
# 6. FINAL SORTING (FOR DASHBOARD DEFAULT VIEW)
# ----------------------------------------------------------

df = df.sort_values("performance_score", ascending=False)

# ----------------------------------------------------------
# 7. SAVE FINAL VALIDATED DATASET
# ----------------------------------------------------------

df.to_csv(OUTPUT_FILE, index=False)

print("\nFinal dataset validated and saved")

# ----------------------------------------------------------
# 8. DASHBOARD LOGIC SUMMARY (PRINT)
# ----------------------------------------------------------

print("\n=== DASHBOARD STRUCTURE ===")

print("""
1. KPI Cards:
   - Total Vendors
   - Top Performers
   - Critical Suppliers
   - Avg Cost
   - Avg Defect

2. Bar Charts:
   - Top 10 Vendors (performance_score)
   - Bottom 10 Vendors

3. Pie Charts:
   - Supplier Category Distribution
   - Risk Distribution

4. Scatter Plot:
   - Cost vs Defect (colored by category)

5. Table:
   - Vendor Name, Score, Category, Decision
""")

print("\nDay 22 Completed Successfully")
```
