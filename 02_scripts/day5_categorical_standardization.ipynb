```python
import pandas as pd
from pathlib import Path
```

```python
PROJECT_ROOT = Path().resolve().parent

INPUT_FILE = PROJECT_ROOT / "01_raw_data/honeyrich_supply_logs.csv"
OUTPUT_FILE = PROJECT_ROOT / "03_processed_data/day5_standardized_supply_data.csv"
```

```python
#load the data
df = pd.read_csv(INPUT_FILE)

print("Dataset Shape:", df.shape)
```

```python
# STANDARDIZE COLUMN NAMES

print("\nStandardizing column names...")

df.columns = (
    df.columns
    .str.strip()
    .str.lower()
    .str.replace(" ", "_")
)

print("Columns after standardization:")
print(df.columns)
```

```python
# CLEAN TEXT COLUMNS
def clean_text(series):

    return(
        series.astype(str)
        .str.strip()
        .str.lower()
        .str.replace(" ", "_")
        .str.replace("-", "_")
    )

text_columns=[
    "vendor_name",
    "product_name",
    "unit"
]

for col in text_columns:

    print(f"\nCleaning column: {col}")

    before = df[col].nunique()

    df[col] = clean_text(df[col])

    after = df[col].nunique()

    print("Unique values before:", before)
    print("Unique values after:", after)
```

```python
# CHECK DUPLICATES
duplicates = df.duplicated().sum()

print("\nDuplicate rows found:", duplicates)
```

```python
# SAVE CLEANED DATA
df.to_csv(OUTPUT_FILE, index=False)

print("\nStandardized dataset saved to:")
print(OUTPUT_FILE)
```
